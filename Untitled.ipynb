# 1. Построение матрицы корреляции для числовых переменных и дальнейшего анализа корреляционных связей между ними.

```python
# Загрузка данных
data = pd.read_csv('german_real_estate_data.csv')

# Выбор числовых переменных для анализа корреляции
numeric_columns = ['square_meters', 'total_floors', 'floor', 'price', 'days_on_market', 'rooms']

# Построение матрицы корреляции
correlation_matrix = data[numeric_columns].corr()

# Визуализация матрицы корреляции с использованием тепловой карты
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

проанализируем результаты нашей корреляционной матрицы:

Квадратные метры (square_meters) не имеют сильной корреляции с другими переменными, что подтверждается значениями корреляции, близкими к нулю.

Количество этажей (total_floors) также слабо коррелирует с другими переменными, хотя есть небольшая положительная корреляция с количеством комнат (rooms).

Этаж (floor) имеет очень слабую корреляцию с остальными переменными.

Цена (price) демонстрирует небольшую положительную корреляцию с квадратными метрами и количеством дней на рынке (days_on_market), что может указывать на некоторые тенденции в рыночной динамике.

Количество дней на рынке (days_on_market) не коррелирует с другими переменными, кроме цены.

Количество комнат (rooms) слабо коррелирует с квадратными метрами и количеством этажей.

# 2. Ящики с усами

```python
# Список категориальных переменных
categorical_columns = ['type', 'city']

# Построение ящиков с усами для каждой категориальной переменной
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=column, y='price', data=data)
    plt.title(f'Boxplot of Price by {column}')
    plt.xlabel(column)
    plt.ylabel('Price')
    plt.xticks(rotation=45)  # Поворот подписей на оси x для лучшей читаемости
    plt.grid(True)
    plt.show()
```

# 3. Графики распределения категориальных переменных

```python
# Группируем данные по типу недвижимости и считаем количество записей в каждой категории
property_type_counts = data['type'].value_counts()

# Строим столбчатую диаграмму
plt.figure(figsize=(8, 6))
property_type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Анализ выбросов и пропущенных значений

1. Визуальный анализ с помощью гистограмм, ящиков с усами и диаграмм рассеяния.
2. Расчет статистических мер, таких как медиана, среднее значение, стандартное отклонение и квартили.
3. Подсчет количества пропущенных значений в каждом столбце.

```python
# Создание сетки графиков для каждого числового столбца
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Построение гистограмм для каждого числового столбца
for ax, col in zip(axes.flatten(), numeric_columns):
    sns.histplot(data=data[col], bins=20, kde=True, ax=ax)
    ax.set_title(f'Histogram of {col}')
    ax.set_xlabel('')
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Построение ящиков с усами для каждого числового столбца
for i, col in enumerate(numeric_columns):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x=data[col], orient='v')
    plt.title(f'Boxplot of {col}')
    plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

```python
# Подсчет пропущенных значений в каждом столбце
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

# 5. Сравнение различных городов и типов недвижимости: сравнить средние значения и распределения различных переменных для разных городов и типов недвижимости.

```python
plt.figure(figsize=(12, 8))

# Построение ящиков с усами для цены по типу недвижимости и городу
plt.subplot(2, 1, 1)
sns.boxplot(x='type', y='price', hue='city', data=data)
plt.title('Comparison of Property Prices by Type and City')
plt.xlabel('Property Type')
plt.ylabel('Price')

# Построение ящиков с усами для квадратных метров по типу недвижимости и городу
plt.subplot(2, 1, 2)
sns.boxplot(x='type', y='square_meters', hue='city', data=data)
plt.title('Comparison of Square Meters by Type and City')
plt.xlabel('Property Type')
plt.ylabel('Square Meters')

plt.tight_layout()
plt.show()
```
